# Construction du MarsCamBot

Notre recommandation pour la construction d'un rover est kit le [CamJam EduKit 3](https://camjam.me/?page_id=1035). Le kit fournit tous les éléments nécessaires à la fabrication d'un rover (moteurs, roues, circuit de contrôle du moteur, avec en prime un capteur de distance, et une breadboard), pour seulement 18 euros hors taxes. Il restera à acquérir une carte Raspberry Pi ainsi qu'une caméra pour construire un rover capable de reconnaissance d'images.

De nombreuses autres posibilités peuvent cependant être envisagées. Le CamJam EduKit semble être une des meilleures en terme de rapport qualité/prix (80 euros par rover, si l'on ajoute la carte Raspberry PI et la caméra). Nous donnons sinon en fin de fiche un aperçu d'alternatives que nous pensons intéressantes si des budgets plus élevés peuvent être considérés.

Pour le chassis, la boîte en carton du CamJam EduKit 3 peut être utilisée. Alternativement, le chassis peut être imprimé en 3D en suivant [ces instructions](https://www.thingiverse.com/thing:1113796).

Le prix minimum est donc d'environ 80 euros (CamJam EduKit 3 + Raspberry Pi 3 + Raspberry Pi Camera + Piles). 

Si le budget le permet, un Raspberry Pi 4 avec l'accélérateur Coral fournira de meilleure performances (reconnaissance d'image jusque 15 images/seconde, soit une réactivité très bonne). Le coût du matériel monte alors à environ 150 euros. Sans l'accélérateur Coral, la reconnaissance d'images sera plutôt de l'ordre de 2 images par seconde, permettant toutefois de réaliser beaucoup d'activités de façon satisfaisante.

## Construction du robot

La construction du robot est très facile. Au plus simple, la boite du CamJam EduKit est utilisée comme chassis. Il s'agit de coller les moteurs sous la boîte, y attacher les roues, et placer la bille à l'avant, grâce à des scotchs fournis dans le kit. Pour le cablâge, les fils des moteurs et la batterie sont reliés au contrôleur moteur, que l'on place sur le Raspberry Pi dans la boîte. La caméra est finalement placée à l'avant. 

La simplicité du kit offre une large place à la créativité pour la construction du chassis. Quelques exemples de rovers construits avec le CamJam EduKit sont données ci-dessous, avec un chassis utilisant la boîte du kit, un chassis imprimé en 3D, ou l'utilisation d'une simple planche de bois.
